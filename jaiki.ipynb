```python
import matplotlib.pyplot as plt
import numpy as np

# Define the vertices of the Gibbs triangle
vertices = np.array([[0, 0], [1, 0], [0.5, np.sqrt(3) / 2]])

# Define the points u and v
u = [0.3, 0.3 * np.sqrt(3) / 2]
v = [0.7, 0.7 * np.sqrt(3) / 2]

# Create the plot
fig, ax = plt.subplots(figsize=(8, 8))

# Plot the triangle
triangle = plt.Polygon(vertices, edgecolor='black', fill=None)
ax.add_patch(triangle)

# Plot points u and v
ax.plot(u[0], u[1], 'ro')  # Point u
ax.plot(v[0], v[1], 'bo')  # Point v

# Annotate the points
ax.text(u[0], u[1], 'u', fontsize=12, ha='right')
ax.text(v[0], v[1], 'v', fontsize=12, ha='left')

# Draw the phase boundary
ax.plot([u[0], v[0]], [u[1], v[1]], 'k--')

# Add an arrow indicating the direction of decreasing temperature
ax.annotate('', xy=v, xytext=u,
            arrowprops=dict(facecolor='black', shrink=0.05))

# Label the vertices
ax.text(vertices[0, 0], vertices[0, 1] - 0.05, 'A', fontsize=12, ha='center')
ax.text(vertices[1, 0], vertices[1, 1] - 0.05, 'B', fontsize=12, ha='center')
ax.text(vertices[2, 0], vertices[2, 1] + 0.05, 'Y', fontsize=12, ha='center')

# Set the plot limits and remove axes
ax.set_xlim(-0.1, 1.1)
ax.set_ylim(-0.1, np.sqrt(3) / 2 + 0.1)
ax.axis('off')

# Display the plot
plt.title("Gibbs Triangle with Phase Boundary and Temperature Direction")
plt.show()
```
